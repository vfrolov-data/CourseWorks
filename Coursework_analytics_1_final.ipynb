```python
# минимальный, максимальный, средний, медианный возраст студентов на каждом курсе
# Step 2.  Применяем агрегацию.
age_by_course.groupby('course_title')['age'].agg(['min', 'max', 'mean', 'median'])
```

```python
# Step 1. Создаём датасет для графика  bar-chart, отражающий количество студентов на каждом курсе. 

bar_chart = main.groupby(['course_title'])['student_id'].nunique()
bar_chart
```

```python
# Step 2 bar-chart, отражающий количество студентов на каждом курсе. Ticks нужно развернуть так, чтобы они были читаемы

plt.bar(bar_chart.index, bar_chart.values)
plt.xticks(rotation=90)

plt.show()
```

```python
# Формирование таблицы для графика   - BEGINING (.unstack() преформатирует датасет и позволяет графику выделять цвет в зависимости от сферы деятельности)
bar_chart_2 = main.groupby(['course_title', 'course_field'])['student_id'].nunique().sort_values().unstack()
bar_chart_2
```

```python
# Формирование таблицы для графика   - STEP 1

bar_chart_2['count'] = bar_chart_2.sum(axis=1)
bar_chart_2
```

```python
# Формирование таблицы для графика   - STEP 2
bar_chart_2.sort_values(by=['count'], inplace=True)
```

```python
# Формирование таблицы для графика   - STEP 3

bar_chart_2.drop(['count'], axis='columns', inplace=True)
bar_chart_2
```

```python
fig, ax = plt.subplots()
plt.style.use('seaborn')
#plt.rcParams.update(plt.rcParamsDefault)


ax = bar_chart_2.plot(kind='barh',  ax=ax, figsize=(16, 8), alpha=0.3, fontsize=12, grid = False, stacked=True) # основной график

plt.vlines(x = bar_chart.median(), ymin=0, ymax=15, linestyle='--',  linewidth=0.75, colors='r', label='Median') # отрисовка медианы - вертикальная линия 



ax.legend( facecolor = 'oldlace',     #  цвет области
           edgecolor = 'r',            #  цвет обрамления
           loc='center right',
           bbox_to_anchor=(1.4, 0.5))           #  смещение легенды

plt.xlabel('Количество студентов', fontsize=10)
plt.ylabel('Название курса',fontsize=10)
plt.title('Количество студентов на каждом курсе')
plt.xticks(rotation=45) # разворачиваем метки по оси Х

# plt.box(False)

plt.tight_layout()  # ajusting plot

plt.show()
```

#### 4.1 (p) Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. Формула: отношение количества студентов, приступивших к выполнению домашнего задания в этом модуле (если дз в модуле несколько, то считать по первому дз в модуле), к количеству студентов, сдавших задание в предыдущем модуле (если дз в модуле несколько, то считать по последнему дз в модуле).

```python
# Выделяем датасет с ДЗ
df_hw = main[main['is_homework'] == True]
df_hw.head(1)
```

```python
# выделяем для каждого курса в каждом модуле номера урокв с первым(min) и последним (max) ДЗ 
df_mix = course_contents[course_contents['is_homework'] == True].groupby(['course_id', 'module_number'])['lesson_number'].agg(['min', 'max']).reset_index()
df_mix.head()
```

```python
# Объединяем df_mix и df_hw с учётом столбца min - нужно получить уроки с первыми ДЗ в модулях
df_min = pd.merge(left = df_mix, right = df_hw, left_on = ['course_id','module_number', 'min'],
                        right_on=['course_id','module_number', 'lesson_number'], how='left')
df_min.head(50)
```

```python
# Объединяем df_mix и df_hw с учётом столбца max - нужно получить уроки с последними ДЗ в модулях
df_max = pd.merge(left = df_mix, right = df_hw, left_on = ['course_id','module_number', 'max'],
                        right_on=['course_id','module_number', 'lesson_number'], how='left')
df_max.head()
```

```python
# ПРОВЕРКА
# Находим модули, где номер урока с первым ДЗ отличается от урока с последним ДЗ в модуле
df_mix[df_mix['min'] != df_mix['max']]
```

```python
# ПРОВЕРКА
df_max[df_max['max'] == 27].head()
```

```python
# Посчитать количество приступивших к ДЗ студентов в каждом модуле
# Считаем общее количество записей с ДЗ - колонка status - в датасете df_min
df_min.groupby(['course_title', 'module_title'])['status'].count()
```

```python
# Выделяем все сданные последние ДЗ (со статусом done) в отделный датасет
df_max_done = df_max[df_max['status'] == 'done']
df_max_done.head()
```

```python
# Посчитать количество студентов, сдавших ДЗ в каждом модуле
# Считаем количество записей с ДЗ со значением done в колонке status в датасете df_max_done
df_max_done.groupby(['course_title', 'module_title'])['status'].count()
```

```python
# размерность датасетов
print(df_min.shape, df_max_done.shape)
```

```python
# Создаём датафрейм с подсчитанным количеством ДЗ (студенты сдавшие ДЗ) в каждом модуле
max_df = df_max_done.groupby(['course_title', 'module_title', 'module_number'])['status'].count().reset_index()
max_df
```

```python
# Создаём датафрейм с подсчитанным количеством ДЗ (студенты приступившие к ДЗ) в каждом модуле
min_df = df_min.groupby(['course_title', 'module_title', 'module_number'])['status'].count().reset_index()
min_df
```

```python
# Переименуем колонку status, чтобы различать в объединнном датасете завершённые и сданные ДЗ

min_df.rename(columns = {'status': 'min_hw'}, inplace=True)
min_df
```

```python
# Переименуем колонку status, чтобы различать в объединнном датасете завершённые и сданные ДЗ
max_df.rename(columns = {'status': 'max_hw'}, inplace=True)
max_df
```

## 5. Метрика успеваемости

Иногда студенты берут курсы, которые оказываются для них неподъемными. Это может быть как по причинам недостаточной изначальной подготовки, так и по причинам, связанным с низкой мотивацией студента. Для улучшения качества контента полезно выявить причину. На основании имеющихся данных придумайте метрику успеваемости студента. **Обоснуйте ее.** Выявите таких студентов на каждом курсе, чтобы предоставить продюсерам список проблемных студентов. 


#### Основная метрика успеваемости студента - процент несделанных домашних заданий. 

#### Несделанное домашнее задание даже при полностью прослушанных лекциях может служить указателем на проблемы с успеваемостью в учёбе. В метрике будем рассматривать домашние задание со статусами fail (несделанное ДЗ) и start/wait - студент начал ДЗ, отправил на проверку, но зачет еще не получил (могут рассматриваться как затягивание со сдачей ДЗ).  

```python
# Part I Вычисление количества ДЗ, которые необходимо сделать на каждом курсе
# Для этого использум изначальный датасет course_contents с планом обучения по каждому курсу

# Создаём датасет на базе course_contents только с ДЗ без обучающих модулей
course_contents_hw = course_contents[course_contents['is_homework'] == True]
course_contents_hw.head()
```

```python
# Вычисляем количество ДЗ, которое необходимо сделать  на каждом курсе
df_contents_hw = course_contents_hw.groupby(['course_id'])['is_homework'].value_counts()
df_contents_hw
```

```python
# Конвертируем в датафрейм (датасет с ДЗ на каждом курсе )
df_hw_bycourse = df_contents_hw.to_frame()
df_hw_bycourse
```

```python
# Убираем мульти-индекс с помощью двойного reset_index
# df_hw_bycourse = df_hw_bycourse.reset_index(0).reset_index(drop=True)
df_hw_bycourse = df_hw_bycourse.reset_index(level=0).reset_index(drop=True)
df_hw_bycourse
```

```python
# Part II Надо создать ДФ, где только студенты со статусом ДЗ fail, start, wait (студенты, у которых проблемы с выполнением ДЗ)

# Создаём датафрейм со строками, где есть домашние задания
df_hw = main[main['is_homework'] == True]
df_hw.head(1)
```

```python
# Создаём датафрейм, где только студенты со статусом ДЗ fail, start, wait
df_hw_undone = df_hw[df_hw['status'] != 'done']
df_hw_undone.head()
```

```python
# Получаем датасет с неуспевающими студентами с разбиением по курсам

# Как показывает работа с фильтром, студент может обучаться на нескольких курсах

by_std_undone = df_hw_undone.groupby(['course_id', 'course_title', 'student_id'])['is_homework'].value_counts()
by_std_undone.head()
```

```python
# Необходимо переименовать колонку, которая содержится в обоих уровнях.
df_std_undone.rename({'is_homework':'is_homework_count'}, axis='columns', inplace=True)
df_std_undone
```

```python
# Убираем мульти-индекс с помощью  reset_index
df_std_undone = df_std_undone.reset_index()
df_std_undone
```

```python
# Удалить ненужную колонку is_homework
df_std_undone.drop('is_homework', axis=1, inplace=True)
```

```python
# Проверка результата
df_std_undone.head()
```

```python
# Объединить датасет с количеством ДЗ на каждом курсе df_hw_bycourse  с датасетом проблемных студентов df_std_undone
# Получаем возможность вычислить отношение несделанных ДЗ каждым студентом к количеству ДЗ на курсе
df_summary = df_std_undone.merge(df_hw_bycourse, how = 'left', left_on=['course_id'], right_on = ['course_id'])
df_summary
```
